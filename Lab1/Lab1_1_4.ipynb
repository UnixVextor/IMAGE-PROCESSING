```python
import numpy as np
import cv2
import math
import matplotlib.image as mpimg


%matplotlib widget
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
img = cv2.imread("image/jumpingFish.jpg")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img = cv2.resize(gray_img,(200, 200))
np_img = np.zeros((gray_img.shape[0], gray_img.shape[1]))

for h in range(gray_img.shape[0]):
    for w in range(gray_img.shape[1]):
        np_img[h][w] = gray_img[h][w]



plt.rcParams["figure.figsize"] = [10, 10]
plt.rcParams["figure.autolayout"] = True

xx,yy = np.mgrid[0:np_img.shape[0],0:np_img.shape[1]]
fig = plt.figure()

ax = fig.gca(projection='3d')
ax.plot_surface(xx,yy,np_img,cmap='gray')

plt.show()
```
