```python
import numpy as np
import cv2
import math
import matplotlib.image as mpimg


%matplotlib widget
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
img = cv2.imread("image/jumpingFish.jpg")
img_rgb = cv2.cvtColor(img,  cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS) 
img_YCrCb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)


file_names = [img_rgb, img_rgb[:,:,0], img_rgb[:,:,1], img_rgb[:,:,2], img_hsv,img_hsv[:,:,0], img_hsv[:,:,1], img_hsv[:,:,2], img_hls, img_hls[:,:,0], img_hls[:,:,1], img_hls[:,:,2],img_YCrCb, img_YCrCb[:,:,0], img_YCrCb[:,:,1], img_YCrCb[:,:,2]]
_, axs = plt.subplots(4,4, figsize=(16,12))
axs = axs.flatten()

for img, ax in zip(file_names, axs):
    # plt.title("abs")
    ax.imshow(img, cmap='gray')
plt.show()
```
