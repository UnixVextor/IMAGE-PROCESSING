```python
import numpy as np
import cv2
import math
import matplotlib.image as mpimg


%matplotlib widget
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
img = cv2.imread("image/jumpingFish.jpg")
rgb_image = np.copy(img)
rgb_image[:,:,0], rgb_image[:,:,2] = img[:,:,2],img[:,:,0] 
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
new_range = np.zeros((gray_image.shape[0], gray_image.shape[1]))
# print(gray_image)

_, axs = plt.subplots(1,2, figsize=(12,4))

def qt(value,bit_depth):
    Qlevel = math.pow(2,bit_depth)
    NormSi = (value - 0) / 255
    Q = math.floor(NormSi * Qlevel)
    return Q

for h in range(gray_image.shape[0]):
    for w in range(gray_image.shape[1]):
        new_range[h][w] = qt(gray_image[h][w],4)
        
        

print(new_range)
plt.imshow(gray_image, cmap='gray')
plt.imshow(new_range, cmap='gray')
plt.show()
```
