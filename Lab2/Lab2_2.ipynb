```python
import numpy as np
import cv2
import math
import matplotlib.image as mpimg


%matplotlib widget
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
img = cv2.imread("images/lab2.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


fig = plt.figure(figsize=(10,7))
row, column = 2,2

img_output = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_output[:,:,0] = cv2.equalizeHist(img_output[:,:,0])
img_output[:,:,1] = cv2.equalizeHist(img_output[:,:,1])
img_output[:,:,2] = cv2.equalizeHist(img_output[:,:,2])

colors = ('r', 'g', 'b')

fig.add_subplot(row,column, 1)
plt.imshow(img_rgb)

fig.add_subplot(row,column, 2)
for i,color in enumerate(colors):
    hist = cv2.calcHist([img_rgb],[i],None,[256],[0,256])
    plt.plot(hist,color = color)


fig.add_subplot(row,column, 3)
plt.imshow(img_output)

fig.add_subplot(row,column, 4)
for i,color in enumerate(colors):
    hist = cv2.calcHist([img_output],[i],None,[256],[0,256])
    plt.plot(hist,color = color)

# plt.imshow(img_output)
```
